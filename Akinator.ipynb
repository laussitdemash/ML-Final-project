```python
import pandas as pd
df = pd.read_csv('modified_akinator_data.csv')
columns_to_convert = df.columns[12:23]
df[columns_to_convert] = df[columns_to_convert].astype(bool)
df['Occupation'].unique()
df['Occupation'] = df['Occupation'].str.replace(r'(formerly)', '').str.replace(r'[1]', '').str.replace(r'[2]', '').str.replace(r'[3]', '').str.replace(r'[4]', '').str.strip(',')
df.loc[df['Hero'] == 'Torbj%C3%B6rn', 'Hero'] = 'Torborn'
df.loc[df['Hero'] == 'L%C3%BAcio', 'Hero'] = 'Lucio'
df['Occupation'] = df['Occupation'].apply(lambda x: ' '.join(x.split()))
# Create exceptions for specific patterns
exceptions = {
    'Professional gamer': 'Professional gamer,',
    'Mech pilot': 'Mech pilot,',
    'MercenaryCEOTalon council member': 'Mercenary, CEO, Talon council member', 
    'Martial artist': 'Martial artist,',
    'Gladiator champion': 'Gladiator champion,',
    'MercenaryHeavy': 'Mercenary, Heavy,',
    'Military leader': 'Military leader,',
    'Shambali monk': 'Shambali monk,',
    'Crusader': 'Crusader,',
    'Overwatch agent': 'Overwatch agent,',
    'Enforcer': 'Enforcer,',
    'Scientist': 'Scientist,',
    'Test subject': 'Test subject,',
    'mechanicBodyguard': 'mechanic Bodyguard',
    'Battle automaton': 'Battle automaton,',
    'Factory worker': 'Factory worker,',
    'Farmhand': 'Farmhand,',
    'Criminal': 'Criminal,',
    'Police officer': 'Police officer,',
    'Soldier': 'Soldier,',
    'Strike commander': 'Strike commander,',
    'Infantrywoman': 'Infantrywoman,',
    'Special Forces operative': 'Special Forces operative,',
    'Overwatch captain': 'Overwatch captain,',
    'Overwatch commander': 'Overwatch commander,',
    'Weapons designer': 'Weapons designer,',
    'AdventurerPilotOverwatch agent,': 'Adventurer, Pilot, Overwatch agent',
    'Ballet dancer': 'Ballet dancer,',
    'Sharpshooter': 'Sharpshooter,',
    'Mechanical engineerAdventurerSquire': 'Mechanical engineer, Adventurer',
    'warriorVigilante': 'warrior, Vigilante',
    'VigilanteShrine': 'Vigilante, Shrine ',
    'ArtistActivistAdventurer': 'Artist, Activist, Adventurer',
    'DJ': 'DJ,',
    'Field medic': 'Field medic,',
    'GeneticistMinister': ' Minister ',
    'GeneticsTalon': 'Genetics, Talon ',
    'Wandering guru': 'Wandering guru,'
}
# Apply exceptions with adjusted regex
for pattern, replacement in exceptions.items():
    df['Occupation'] = df['Occupation'].str.replace(pattern, replacement)

# Remove comma after 'Soldier'
df.loc[df['Occupation'] == 'Soldier,', 'Occupation'] = 'Soldier'
# Remove extra comma after 'Mech pilot'
df.loc[df['Occupation'].str.contains('Test subject, Mech pilot,'), 'Occupation'] = 'Test subject, Mech pilot, mechanic Bodyguard'
df = df.drop(['Real Name', 'Age', 'Health', 'Armor' ], axis=1)
df['Nationality'] = df['Nationality'].str.replace(r'[2]', '')
df['Nationality'] = df['Nationality'].replace('0', 'Robot(No Nationality)')
# Create dummy variables for 'Nationality'
df_nationality_dummies = pd.get_dummies(df['Nationality'], prefix='Nationality')
df = pd.concat([df, df_nationality_dummies], axis=1)
df.drop('Nationality', axis=1, inplace=True)

# Create dummy variables for 'Occupation'
df_occupation_dummies = pd.get_dummies(df['Occupation'], prefix='Occupation')
df = pd.concat([df, df_occupation_dummies], axis=1)
df.drop('Occupation', axis=1, inplace=True)

# Create dummy variables for 'Weapons'
df_weapons_dummies = pd.get_dummies(df['Weapons'], prefix='Weapons')
df = pd.concat([df, df_weapons_dummies], axis=1)
df.drop('Weapons', axis=1, inplace=True)

# Create dummy variables for 'Abilities(LSHIFT)'
df_lshift_dummies = pd.get_dummies(df['Abilities(LSHIFT)'], prefix='Abilities_LSHIFT')
df = pd.concat([df, df_lshift_dummies], axis=1)
df.drop('Abilities(LSHIFT)', axis=1, inplace=True)

# Create dummy variables for 'Abilities(E)'
df_e_dummies = pd.get_dummies(df['Abilities(E)'], prefix='Abilities_E')
df = pd.concat([df, df_e_dummies], axis=1)
df.drop('Abilities(E)', axis=1, inplace=True)

df = pd.get_dummies(df, columns=['Role'])

# Create dummy variables for 'Abilities(Q)'
df_q_dummies = pd.get_dummies(df['Abilities(Q)'], prefix='Abilities_Q')
df = pd.concat([df, df_q_dummies], axis=1)
df.drop('Abilities(Q)', axis=1, inplace=True)
df
```

```python
from sklearn.tree import DecisionTreeClassifier
from random import randint

class Akinator:
    def __init__(self, df):
        self.df = df
        self.random_state = randint(1, 42)
        self.clf = DecisionTreeClassifier(random_state=self.random_state)
        self.build_decision_tree()
        self.asked_features = set()

    def build_decision_tree(self):
        X = self.df.drop(['Hero'], axis=1)
        y = self.df['Hero']
        self.clf.fit(X, y)
    
    def ask_question(self, feature):
        answer = input(f"Is your {feature.replace('_', ' ')} (yes/no/idk): ").lower()
        while answer not in ['yes', 'no', 'idk']:
            answer = input("Invalid input. Please enter 'yes', 'no', or 'idk': ").lower()
        return answer

    def display_result(self, most_likely_hero):
        print(f"I think the hero is {most_likely_hero}.")
    
    def confirm_hero(self):
        answer = input("Is your hero the same as the predicted hero? (yes/no): ").lower()
        while answer not in ['yes', 'no']:
            answer = input("Invalid input. Please enter 'yes' or 'no': ").lower()

        if answer == 'yes':
            print("Good :)")
            return True
        else:
            return False

    def get_unasked_feature(self):
        # Get a feature that has not been asked before
        unasked_features = set(self.df.columns[1:]) - self.asked_features

        if unasked_features:
            return unasked_features.pop()
        else:
            return None
        

    def play(self, node_index=0):
        node = self.clf.tree_
        feature_names = self.df.columns[1:] 
        while node.feature[node_index] != -2:
            feature_index = node.feature[node_index]
            feature = feature_names[feature_index]
            # Check if the feature has been asked before
            if feature not in self.asked_features:
                self.asked_features.add(feature)
            else:
                # Get a new unasked feature
                feature = self.get_unasked_feature()
                if not feature:
                    print("I've asked all available questions. Unable to proceed.")
                    return
                
    
            question = f"hero {feature} ?"
            answer = self.ask_question(question)
    
            if answer == "idk":
                node_index += 1
            elif answer == 'yes':
                node_index = node.children_right[node_index]
            elif answer == 'no':
                node_index = node.children_left[node_index]
    
        leaf_value = node.value[node_index]
        predicted_class = int(leaf_value.argmax())
        predicted_hero = self.clf.classes_[predicted_class]
        self.display_result(predicted_hero)
        if self.confirm_hero():
            return

        # Continue guessing if the confirmed hero is not correct.
        self.play()

         
akinator = Akinator(df.copy())
akinator.play()
```
