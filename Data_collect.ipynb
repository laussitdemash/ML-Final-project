```python
import requests
from bs4 import BeautifulSoup

# URL of the Overwatch Fandom page
url = "https://overwatch.fandom.com/wiki/Heroes"

# Heroes dictionary
heroes = {
    "Tank": ["D.Va", "Doomfist", "Junker Queen", "Mauga", "Orisa", "Ramattra", "Reinhardt", "Roadhog", "Sigma", "Winston", "Wrecking Ball", "Zarya"],
    "Damage": ["Ashe", "Bastion", "Cassidy", "Echo", "Genji", "Hanzo", "Junkrat", "Mei", "Pharah", "Reaper", "Sojourn", "Soldier: 76", "Sombra", "Symmetra", "Torbjörn", "Tracer", "Widowmaker"],
    "Support": ["Ana", "Baptiste", "Brigitte", "Illari", "Kiriko", "Lifeweaver", "Lúcio", "Mercy", "Moira", "Zenyatta"]
}

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Create a dictionary to store hero links
    hero_links = {}

    # Update the existing dictionary with the additional heroes
    heroes.update({
        "Junker Queen": "Junker Queen",
        "Wrecking Ball": "Wrecking Ball",
        "Soldier: 76": "Soldier: 76"
    })

    # Iterate through the roles and heroes in the dictionary
    for role, heroes_list in heroes.items():
        for hero_name in heroes_list:
            # Find the hero link on the page using the modified name
            hero_element = soup.find('a', {'title': hero_name.replace(" ", "_")})

            # If the hero link is not found, try with the original name
            if not hero_element:
                hero_element = soup.find('a', {'title': hero_name})

            # If the hero link is found, add it to the dictionary
            if hero_element:
                hero_links[hero_name] = "https://overwatch.fandom.com" + hero_element['href']

    # Print the hero names and links
    for hero, link in hero_links.items():
        print(f"{hero}: {link}")

else:
    print(f"Failed to retrieve the page. Status Code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

data = [
    ["D.Va", "https://overwatch.fandom.com/wiki/D.Va"],
    ["Doomfist", "https://overwatch.fandom.com/wiki/Doomfist"],
    ["Junker Queen", "https://overwatch.fandom.com/wiki/Junker_Queen"],
    ["Mauga", "https://overwatch.fandom.com/wiki/Mauga"],
    ["Orisa", "https://overwatch.fandom.com/wiki/Orisa"],
    ["Ramattra", "https://overwatch.fandom.com/wiki/Ramattra"],
    ["Reinhardt", "https://overwatch.fandom.com/wiki/Reinhardt"],
    ["Roadhog", "https://overwatch.fandom.com/wiki/Roadhog"],
    ["Sigma", "https://overwatch.fandom.com/wiki/Sigma"],
    ["Winston", "https://overwatch.fandom.com/wiki/Winston"],
    ["Wrecking Ball", "https://overwatch.fandom.com/wiki/Wrecking_Ball"],
    ["Zarya", "https://overwatch.fandom.com/wiki/Zarya"],
    ["Ashe", "https://overwatch.fandom.com/wiki/Ashe"],
    ["Bastion", "https://overwatch.fandom.com/wiki/Bastion"],
    ["Cassidy", "https://overwatch.fandom.com/wiki/Cassidy"],
    ["Echo", "https://overwatch.fandom.com/wiki/Echo"],
    ["Genji", "https://overwatch.fandom.com/wiki/Genji"],
    ["Hanzo", "https://overwatch.fandom.com/wiki/Hanzo"],
    ["Junkrat", "https://overwatch.fandom.com/wiki/Junkrat"],
    ["Mei", "https://overwatch.fandom.com/wiki/Mei"],
    ["Pharah", "https://overwatch.fandom.com/wiki/Pharah"],
    ["Reaper", "https://overwatch.fandom.com/wiki/Reaper"],
    ["Sojourn", "https://overwatch.fandom.com/wiki/Sojourn"],
    ["Soldier: 76", "https://overwatch.fandom.com/wiki/Soldier:_76"],
    ["Sombra", "https://overwatch.fandom.com/wiki/Sombra"],
    ["Symmetra", "https://overwatch.fandom.com/wiki/Symmetra"],
    ["Torbjörn", "https://overwatch.fandom.com/wiki/Torbj%C3%B6rn"],
    ["Tracer", "https://overwatch.fandom.com/wiki/Tracer"],
    ["Widowmaker", "https://overwatch.fandom.com/wiki/Widowmaker"],
    ["Ana", "https://overwatch.fandom.com/wiki/Ana"],
    ["Baptiste", "https://overwatch.fandom.com/wiki/Baptiste"],
    ["Brigitte", "https://overwatch.fandom.com/wiki/Brigitte"],
    ["Illari", "https://overwatch.fandom.com/wiki/Illari"],
    ["Kiriko", "https://overwatch.fandom.com/wiki/Kiriko"],
    ["Lifeweaver", "https://overwatch.fandom.com/wiki/Lifeweaver"],
    ["Lúcio", "https://overwatch.fandom.com/wiki/L%C3%BAcio"],
    ["Mercy", "https://overwatch.fandom.com/wiki/Mercy"],
    ["Moira", "https://overwatch.fandom.com/wiki/Moira"],
    ["Zenyatta", "https://overwatch.fandom.com/wiki/Zenyatta"]
]
# Create a dictionary from the data list
hero_links = dict(data)

# Create a list to store hero data as dictionaries
hero_data_list = []

# Function to scrape hero information from a given URL
def scrape_hero_info(hero_url):
    response = requests.get(hero_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting specific information based on HTML structure
        real_name_element = soup.find('div', string='Real Name')
        real_name_value = real_name_element.find_next('td').text.strip() if real_name_element else "Not available"

        age_element = soup.find('div', string='Age')
        age_value = age_element.find_next('td').text.strip() if age_element else "Not available"

        nationality_element = soup.find('div', string='Nationality')
        nationality_value = nationality_element.find_next('td').text.strip() if nationality_element else "Not available"

        occupation_element = soup.find('div', string='Occupation')
        occupation_value = occupation_element.find_next('td').text.strip() if occupation_element else "Not available"

        role_element = soup.find('div', string='Role')
        role_value = role_element.find_next('td').text.strip() if role_element else "Not available"

        health_element = soup.find('div', string='Health')
        health_value = health_element.find_next('td').text.strip() if health_element else "Not available"

        armor_element = soup.find('div', string='Armor')
        armor_value = armor_element.find_next('td').text.strip() if armor_element else "Not available"

        # Create a dictionary for the current hero
        hero_data = {
            "Hero": hero_url.split("/")[-1],  # Extract hero name from the URL
            "Real Name": real_name_value,
            "Age": age_value,
            "Nationality": nationality_value,
            "Occupation": occupation_value,
            "Role": role_value,
            "Health": health_value,
            "Armor": armor_value,
            "Weapons": "",  # Empty for now
            "Abilities": ""  # Empty for now
        }

        hero_data_list.append(hero_data)

    else:
        print(f"Failed to retrieve the page for {hero_url}. Status Code: {response.status_code}")

# Loop through each hero link and scrape information
for hero, link in hero_links.items():
    print(f"Scraping information for {hero}...")
    scrape_hero_info(link)

# Create a DataFrame from the list of hero data
df = pd.DataFrame(hero_data_list)

# Save the DataFrame to a CSV file
df.to_csv('hero_data.csv', index=False)

print("Hero data saved to hero_data.csv")
```

```python
import re
def scrape_weapons(hero_url):
    response = requests.get(hero_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the element containing "Key mouse 1"
        key_mouse_1_element = soup.find('img', alt=re.compile(r'Key mouse 1', re.IGNORECASE))

        if key_mouse_1_element:
            # Navigate to the parent element and extract the weapon name
            weapon_element = key_mouse_1_element.find_parent('div', class_='abilityHeader')
            weapon_name = weapon_element.text.strip() if weapon_element else "Not available"
        else:
            weapon_name = "Not available"

        # Update the DataFrame with the scraped weapon information
        hero_name = hero_url.split("/")[-1]
        df.loc[df['Hero'] == hero_name, 'Weapons'] = str(weapon_name)


    else:
        print(f"Failed to retrieve the page for {hero_url}. Status Code: {response.status_code}")

# Loop through each hero link and scrape weapons information
for hero, link in hero_links.items():
    print(f"Scraping weapons information for {hero}...")
    scrape_weapons(link)

# Save the updated DataFrame to a CSV file
df.to_csv('hero_data_updated.csv', index=False)

print("Updated hero data saved to hero_data_updated.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load the existing CSV file
df = pd.read_csv('hero_data_updated.csv')

# Function to scrape LSHIFT ability information from a given URL
def scrape_lshift_ability_info(ability_url):
    response = requests.get(ability_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting specific information based on HTML structure
        lshift_ability_element = soup.find('span', {'class': 'keybind'}, string='LSHIFT')
        
        if lshift_ability_element:
            ability_name_element = lshift_ability_element.find_parent().find_parent().find_previous('div', {'class': 'abilityHeader'})
            ability_name = ability_name_element.text.strip() if ability_name_element else "Not available"

            description_element = lshift_ability_element.find_parent().find_parent().find_next('div', {'style': 'vertical-align:top; font-size:85%;'})
            description = description_element.text.strip() if description_element else "Not available"

            return f"{ability_name} - {description}"

        else:
            return "LSHIFT ability not found"

    else:
        return f"Failed to retrieve the page for {ability_url}. Status Code: {response.status_code}"

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    hero_name = row['Hero']
    hero_url = row['Hero'].replace(" ", "_")

    # Construct the hero's abilities URL
    abilities_url = f"https://overwatch.fandom.com/wiki/{hero_url}#Abilities"

    # Scrape LSHIFT ability information using the function
    lshift_ability_info = scrape_lshift_ability_info(abilities_url)

    # Update the corresponding row in the DataFrame
    df.at[index, 'Abilities'] = str(lshift_ability_info)


# Save the updated DataFrame to the CSV file
df.to_csv('hero_data_updated.csv', index=False)

print("LSHIFT Abilities data updated and saved to hero_data_updated.csv")
```
